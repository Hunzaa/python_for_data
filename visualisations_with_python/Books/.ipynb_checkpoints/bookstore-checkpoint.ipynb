## 5. What are people reading the most?

##### I wanted to find the books that got the most attention. Not necessarily the highest rated, but the ones that the "most" people actually rated.

```python
# most rated boooks from RATING table
rated_books = ratings[ratings["Book-Rating"] > 0]
highest_rated_books = rated_books.groupby("ISBN")["Book-Rating"].count() # group by ISBN and count how many times each book was rated
highest_rated_books = highest_rated_books.sort_values(ascending=False) # sort by count

print(highest_rated_books.head(5))
```

##### I realised using only the ratings table wasn’t enough because it only had ISBNs.
##### So I joined it with the books table to get more useful details like the book title, author, and image.

### Top 5 Highest Rated Books (with Book Title, Author, and Image)

```python
# doing the same thing but joining details from Book table with Image
rated_books = ratings[ratings["Book-Rating"] > 0]
highest_rated_books = rated_books.groupby("ISBN")["Book-Rating"].count()
highest_rated_books = highest_rated_books.sort_values(ascending=False)

highest_rated_books = highest_rated_books.reset_index() # reseting isbn from an idex to a column
highest_rated_books = highest_rated_books.merge(books[["ISBN", "Book-Title", "Book-Author", "Image-URL-M"]], 
                                              on="ISBN", how="left") # joining
highest_rated_books = highest_rated_books.rename(columns={"Book-Rating": "NoOfRatings"}) #rename

print(highest_rated_books.head(5) )
```

##### I noticed that some books show up with NaN in the title, author, and image. That’s because those ISBNs exist in the ratings file but not in the books file. Might be bacasue the book is not in the bookstore anymore? 

##### It kind of makes the book reduntant because we dont really know what the Book title is. Need to clean again...

#### Cleaning and Checking

```python
# checking
print("NaN in Highest Rated Books:", highest_rated_books.isna().sum().sum(), "missing values\n")
print(highest_rated_books.isna().sum()) 
```

```python
# cleaning 
# removing rows with missing title, author or image
highest_rated_books = highest_rated_books.dropna(subset=["Book-Title", "Book-Author", "Image-URL-M"])
print(highest_rated_books.isna().sum()) 
```

##### Now that I’ve checked and cleaned the table, I can go ahead with what I originally planned.
##### Below I’ve created a clean visual of the top 5 books that’s easy to read and gives insight at a glance.

```python
# creating a visual for top 5

top5 = highest_rated_books.head(5)

# i  wanted to show a clean table table but couldn't do an actual table 
# so I made a small blank space (8 wide, 2 tall) 
# and hid all the axis stuff (lines, ticks, numbers) to keep it neat
fig, ax = plt.subplots(figsize=(8, 2))
ax.axis('off')

# creating the table 
# converting rows into a plain list of lists to use in table
table_data = top5[["Book-Title", "Book-Author", "NoOfRatings"]].values.tolist() 
column_labels = ["Book Title", "Author", "No. of Ratings"]

# customising my table 
table = ax.table(cellText=table_data, colLabels=column_labels, loc='center', cellLoc='left')
table.scale(1, 2)
table.auto_set_font_size(False)
table.set_fontsize(9)

# customising title
plt.title("Top 5 Most Rated Books", fontsize=12, pad=30, color="red")
plt.tight_layout()
plt.show()
```

## 6. Who's Reading? Age Groups & Countries

##### Next I wanted to identify the readers, like: 
##### - Where are they from?
##### - How old are they? 

##### I will look at users age and country.

```python
# dont have country column so
# from the users table creating a new column called Country
# by taking the location and splitting it by commas  
# and then grabbing the last part (country) while revoming sapces 
users["Country"] = users["Location"].apply(lambda Location: Location.split(",")[-1].strip())
users.head(5)
```

### Where are most readers from?

```python
# looking at the top countries by number of users
top_countries = users["Country"].value_counts().head(10)
print(top_countries)
```

##### After looking at the count I relaized that there are some blanks (some users didn’t have a country listed). 
##### So need to clean them them to avoid any vague info on charts. 

```python
# cleaning to handle null Countries
users["Country"] = users["Country"].replace("", "Unknown") # placeholder
print("Empty countries:", (users["Country"] == "").sum())
```

##### Now that the country data is cleaned, I’m continuing to create a chart.

```python
# bar chart 
# top 10 counties by user count
top_countries = users["Country"].value_counts().head(10)
top_countries = top_countries.reset_index()
top_countries.columns = ["Country", "Readers"]

sns.barplot(x="Readers", y="Country", data=top_countries, errorbar=None, color="orchid")
plt.title("Top 10 Countries by Readers")
plt.xlabel("Readers")
plt.ylabel("Country")
plt.show()
```

##### This bar chart gives us a quick visual of the top 10 countries with the highest number of readers. You can see that most reader are from USA, followed by Canada and the UK.

### How old are they? (Age groups and their preferred Authors)

##### First, I wanted to split readers into age groups, like teens, young adults, middle aged.

```python
# creating a new column in USERS called AgeGroup
# putting ages into 6 categories/bins
users["AgeGroup"] = pd.cut(users["Age"], 
                           bins=[0, 18, 25, 35, 50, 70, 100], 
                           labels=["<18", "18–25", "26–35", "36–50", "51–70", "70+"],
                           right=False)
# print(users.head(3), "\n\n")
```

##### Next, I created a new combined table by merging all three datasets (ratings, users, and books).
##### This helped bring everything into one place 
##### - the user ID, their age group from USERS
##### - the book they rated from RATINGS
##### - and the author of that book from BOOKS

```python
# creating a table that will have information merged from all tables
merge_all_tables = ratings.merge(users[["User-ID", "Age", "AgeGroup"]], on="User-ID").merge(books[["ISBN", "Book-Author"]], on="ISBN", how="left")
# print(merge_all_tables.head(3), "\n\n")
```

##### Now that everything was merged, I wanted to know which authors are most popular in each age group?

```python
# grouped age group and author
# counted how many ratings each author got from that age group
author_counts = merge_all_tables.groupby(["AgeGroup", "Book-Author"], observed=False)["Book-Rating"].count().reset_index()
#print(author_counts.head(5), "\n\n")

# sorting author counts by highest to lowest
top_authors = author_counts.sort_values(["AgeGroup", "Book-Rating"], ascending=[True, False]) \
                           .groupby("AgeGroup", observed=False).head(2)

print(top_authors, "\n\n")
```

```python
# top authors by age group
# Show it in a bar chart
sns.barplot(data=top_authors, x="AgeGroup", y="Book-Rating", hue="Book-Author", palette="Set2")
plt.title("Top Two Authors by Age Group")
plt.tight_layout()
plt.show()
```

##### Turns out:
##### - R. L. Stine is a hit with younger readers,
##### - Stephen King and Nora Roberts is a favourite for most of the age groups


##### Although this chart was mainly made to show the top two authors for each age group, it actually also give us a rough idea of which age groups are more active. 
##### - Readers aged 26–35 seem the most engaged (tallest bars).
##### - The youngest and oldest age groups are the least actibe, which could mean they either don’t rate much or don’t read as often.

## 7. Do Different Countries Love the Same Books?

```python
# merge all tables
merged = ratings.merge(users, on="User-ID").merge(books[["ISBN", "Book-Title"]], on="ISBN", how="left")
# print(merged.head(3), "\n\n")
```

##### Focusing on top 10 countries and top 10 books to keep things clean and not too crowded.

```python
# getting top 10 countries with most readers
# getting top 10 most rated books
top_countries = merged["Country"].value_counts().head(10).index
top_books = merged["Book-Title"].value_counts().head(10).index

# keep only rows with top countries AND top books
top10 = merged[merged["Country"].isin(top_countries) & merged["Book-Title"].isin(top_books)]
# print(top10)
```

##### Getting the Average Rating Per Country Per Book
##### I want to know how people in each country(top10) rated each book(top10) on average. 

```python
# grouping by country and book, and calculating average rating
group_top10 = top10.groupby(["Country", "Book-Title"])["Book-Rating"].mean().reset_index()
# print(group_top10)
```

##### Heatmaps need a grid format so reshaping:
##### rows = countries
##### columns = books
##### values = average rating

```python
# reshaping the table for heatmap
heatmap_table = group_top10.pivot(index="Country", columns="Book-Title", values="Book-Rating")
```

##### Finally, creating the Heatmap to spot patterns across countries and books.

```python
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_table, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Average Ratings by Country and Book (Top 10)")
plt.xlabel("Book Title")
plt.ylabel("Country")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

1. Some books are international hits. The Da Vinci Code and The Lovely Bones were rated fairly well across most countries. 
2. The Nanny Diaries got a perfect 10 in Spain — but dropped down to around 3 in the UK. Same book, totally different vibes, so not everyone agrees. 
3. France really loved “A Painted House” with an 8, while most other countries gave it just a 2 or 3.
4. So tastes clearly shift depending on where you're from.
5. There are some strange results too — like 0s coming nan or unknown locations. Probably just some messy data sneaking in.
6. USA and UK are quiet balanced. They sit mostly in the middle for every book (no extreme lows or highs).

This chart was fun way to see how book taste can vary across borders. What’s a bestseller in one country might be a meh in another.

## 8. Book Recommeder
##### Confused on what book to read? Try my recommender.
(P.S. Doesn't work on GitHub)

```python
# joining ratings and books tables
merged2 = ratings.merge(books[["ISBN", "Book-Title", "Book-Author"]], on="ISBN", how="left")

# Filter only meaningful ratings
# merged2 = merged2[merged2["Book-Rating"] > 0]
# print(merged2.isna().sum()) 

# many rows had missing author and title (because some ISBN's in ratings are not in boooks table)
merged2 = merged2.dropna(subset=["Book-Author"])
#print(merged2.isna().sum()) 

# questions
favourite_author = input("Who's your favourite author? ") #Nora Roberts
min_rating = int(input("Minimum rating you'd consider (1-10)? "))

# recommendations
recommendations = merged2[merged2["Book-Author"].str.contains(favourite_author)]
recommendations = recommendations.groupby(["Book-Title", "Book-Author"])["Book-Rating"].mean().reset_index()
recommendations = recommendations[recommendations["Book-Rating"] >= min_rating]
recommendations = recommendations.sort_values(by="Book-Rating", ascending=False)

# show top 5 recommendatosns
print("\nRecommended Books for You:\n")
print(recommendations.head(5))
```
